```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('global-warming-annual-temperature-anomaly.csv', encoding = 'windows-1251')
```

```python
df.head()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Year', y='Global average temperature anomaly relative to 1861-1890')
plt.title('Глобальные температурные аномалии (1880–2025)')
plt.xlabel('Год')
plt.ylabel('Отклонение температуры (°C)')
plt.grid(True)
plt.savefig('temperature_anomaly_trend.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
df['decade'] = (df['Year'] // 10) * 10
decade_avg = df.groupby('decade')['Global average temperature anomaly relative to 1861-1890'].mean().reset_index()
```

```python
sns.barplot(data=decade_avg, x='decade', y='Global average temperature anomaly relative to 1861-1890', hue='decade', palette='coolwarm', legend=False)
plt.title('Средняя температурная аномалия по десятилетиям')
plt.ylabel('Среднее отклонение (°C)')
plt.xlabel('Десятилетия')
plt.xticks(rotation=45)
plt.savefig('temperature_anomaly_decade.png', dpi=300, bbox_inches='tight')
plt.show()
```
